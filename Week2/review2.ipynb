```python
# 세트 요소 삭제
engineers = set(['Kim','Park','Lee'])
programmers = set(['Kim','Song','Choi'])
managers = set(['Chun','Seo','Oh'])

for group in [engineers, programmers, managers]: 
	group.discard('Kim') 		# 모든 그룹에서 “Kim"을 삭제한다.  
	print(group)
```

```python
 # 부분 집합 연산
A = {"apple", "banana", "grape"}
B = {"apple", "banana", "grape", "kiwi"}

if A < B : 			# 또는 A.issubset(B)
	print("A는 B의 부분 집합입니다.")
 
A.issubset(B)
#B.issubset(A)
```

```python
# ==, != 연산
A = {"apple", "banana", "grape"}
B = {"apple", "banana", "grape"}

if A == B :
	print("A와 B는 같습니다.")
else :
	print("A와 B는 같지 않습니다.")
```

```python
# 합집합, 교집합, 차집합
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

print(set(list1) | set(list2))
print(set(list1) & set(list2))
print(set(list1) - set(list2))

print(set(list1).union(set(list2)))
print(set(list1).intersection(set(list2)))
print(set(list1).difference(set(list2)))
```

```python
# 리스트 <=> 세트
list1 = [1, 2, 3, 4, 5, 1, 2, 4]
len(set(list1))
```

```python
# 공통 정수 찾기
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
set(list1) & set(list2)
```

```python
# Lab: 문자열의 공통 문자 찾기
s1 = input("첫 번째 문자열:")
s2 = input("두 번째 문자열:")

list1 = list(set(s1) & set(s2))		# 세트로 만들고 교집합 연산 
print(list1)

print("\n공통적인 글자:", end=" ")
for i in list1:
    print(i, end=" ")
```

```python
# Lab: 중복되지 않은 단어의 개수 세기
txt = input("입력 텍스트: ")
words = txt.split(" ")
print(words)

unique = set(words)			# 집합으로 만들면 자동적으로 중복 제거 
print(unique)
print("사용된 단어의 개수= ", len(unique))
```

```python
# 단어 빈도 계산 1
text_data ="Create the highest, grandest vision possible for your life, because you become what you believe"

word_dic = {}				    # 단어의 출현 횟수를 저장하는 딕셔너리를 생성
for w in text_data.split():		# 텍스트를 단어로 분리하여 반복함 
     if w in word_dic:			# 단어가 이미 딕셔너리에 있으면
         word_dic[w]  += 1  	# 출현 횟수를 1 증가한다. 
     else:				        # 처음 나온 단어이면 1로 초기화 
         word_dic[w]   = 1  
print(word_dic)

for w, count in sorted(word_dic.items()):	# 키와 값을 정렬하여 반복 처리한다. 
     print(w, "의 등장횟수=", count)
```
