##### Lab: BMI 계산

```python
import numpy as np
heights = [1.83, 1.76, 1.69, 1.86, 1.77, 1.73]      # 여기까지는 그냥 python array
weights = [ 86,   74,   59,   95,   80,   68 ]

np_heights = np.array(heights)
np_weights = np.array(weights)

bmi = np_weights/(np_heights**2)
print(bmi)

print(bmi[bmi > 25])   # BMI가 25를 넘는 사람의 BMI만을 출력
```

##### Lab: 평균과 중간값 계산하기

```python
import numpy as np 
 
players = np.zeros((100, 3))
#print(players)
players[:, 0] = 10 * np.random.randn(100) + 175 
players[:, 1] = 10 * np.random.randn(100) + 70
players[:, 2] = np.floor(10 * np.random.randn(100)) + 22  # 나이는 정수로
#players[:, 2] = 10 * np.random.randn(100) + 22
#print(players)

heights = players[:, 0] 
print('신장 평균값:', np.mean(heights))
print('신장 중앙값:', np.median(heights))

weights = players[:, 1] 
print('체중 평균값:', np.mean(weights))
print('체중 중앙값:', np.median(weights))

ages = players[:, 2] 
print('나이 평균값:', np.mean(ages))
print('나이 중앙값:', np.median(ages))
```
